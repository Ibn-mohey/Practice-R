```r
## Find a monitor for New York State that exists in both datasets
# get New york with the state code 
site0 <- unique(subset(pm0, State.Code == 36, c(County.Code, Site.ID))) 
site1 <- unique(subset(pm1, State.Code == 36, c(County.Code, Site.ID)))
```

```r
#what is exist in the both
both <- intersect(site0, site1)
print(both)
```

```r
## Find how many observations available at each monitor
pm0$county.site <- with(pm0, paste(County.Code, Site.ID, sep = ".")) #county + site ID 
pm1$county.site <- with(pm1, paste(County.Code, Site.ID, sep = ".")) #county + site ID 

```

```r
## Show state-wide means and make a plot showing trend
mn0 <- with(pm0, tapply(Sample.Value, State.Code, mean, na.rm = T))
str(mn0)
summary(mn0)
```

```r
mn1 <- with(pm1, tapply(Sample.Value, State.Code, mean, na.rm = T))
str(mn1)
summary(mn1)
```

```r
## Make separate data frames for states / years
d0 <- data.frame(state = names(mn0), mean = mn0)
d1 <- data.frame(state = names(mn1), mean = mn1)
mrg <- merge(d0, d1, by = "state")
dim(mrg)
head(mrg)
```

```r
## Connect lines
par(mfrow = c(1, 1))
with(mrg, plot(rep(1999, 52), mrg[, 2], xlim = c(1998, 2013)))
with(mrg, points(rep(2012, 52), mrg[, 3]))
segments(rep(1999, 52), mrg[, 2], rep(2012, 52), mrg[, 3])
```
